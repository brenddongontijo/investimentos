```python
import requests
import numpy
import pandas
```

# Coleta de dados

```python
url = "https://www.fundsexplorer.com.br/ranking"
headers = {
    'User-Agent':
        'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:98.0) Gecko/20100101 Firefox/98.0'
}
```

```python
response = requests.get(url=url, headers=headers)
if response.status_code == 200:
    data_frame = pandas.read_html(response.content, encoding='utf-8')[0]
```

# Tratamento de Dados

# Definindo tipo de Dados Categóricos e Numéricos

```python
categorical_columns = ['Códigodo fundo', 'Setor']
data_frame[categorical_columns] = data_frame[categorical_columns].astype('category')
```

```python
float_columns = list(data_frame.iloc[:, 2:-1].columns)
data_frame[float_columns] = data_frame[float_columns].fillna(value=0)
data_frame[float_columns] = data_frame[float_columns].applymap(lambda x: str(x).replace('R$','').replace('.','').replace('%','').replace(',', '.'))
data_frame[float_columns] = data_frame[float_columns].astype('float')
```

# Realizando ajustes finais no dataframe

```python
# checking infinity values
data_frame[numpy.isinf(data_frame[float_columns]).any(1)]
```

```python
# adjusting columns
data_frame['Liquidez Diária'] = data_frame['Liquidez Diária']/10
data_frame['P/VPA'] = data_frame['P/VPA']/1000
```

```python
indicadores = [
    'Códigodo fundo',
    'Setor',
    'DY (12M)Acumulado',
    'VacânciaFísica',
    'P/VPA',
    'QuantidadeAtivos',
    'Liquidez Diária',
    'PatrimônioLíq.',
]
```

```python
df_aux = data_frame[indicadores]
```

```python
media_setor = df_aux.groupby('Setor').agg(['mean', 'std'])
```

```python
def oportunidade_media_setor(dataframe, setor='Shoppings', label_setor='Setor'):
    media_setor = df_aux.groupby('Setor').agg(['mean', 'std'])
    
    df_setor = dataframe[dataframe[label_setor].isin([setor])]
    
    filter = \
            (df_setor['QuantidadeAtivos'] > 0) & \
            (df_setor['Liquidez Diária'] > 4000) & \
            (df_setor['P/VPA'] < 1.0) & \
            (df_setor['DY (12M)Acumulado'] > media_setor.loc[setor, ('DY (12M)Acumulado', 'mean')])
            
    print('Média do setor Yield: {}'.format(media_setor.loc[setor, ('DY (12M)Acumulado', 'mean')]))
    print('Média do setor P/VPA: {}'.format(media_setor.loc[setor, ('P/VPA', 'mean')]))
    print('Média do setor Ativos: {}'.format(media_setor.loc[setor, ('QuantidadeAtivos', 'mean')]))
    
    return df_setor[filter]
```

```python
oportunidade = oportunidade_media_setor(df_aux, setor='Shoppings')
oportunidade.sort_values('DY (12M)Acumulado', ascending=False, inplace=True)
# oportunidade['PatrimônioLíq.'] *= 10
oportunidade['PatrimônioLíq.'].apply(lambda x: '%.1f' % x).values.tolist()
oportunidade
```
