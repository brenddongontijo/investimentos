```python
import requests
import numpy
import pandas
```

# Coleta de dados

```python
url = "https://www.fundsexplorer.com.br/ranking"
headers = {
    'User-Agent':
        'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:98.0) Gecko/20100101 Firefox/98.0'
}
```

```python
response = requests.get(url=url, headers=headers)
if response.status_code == 200:
    data_frame = pandas.read_html(response.content, encoding='utf-8')[0]
```

# Tratamento de Dados

# Definindo tipo de Dados Categóricos e Numéricos

```python
categorical_columns = ['Códigodo fundo', 'Setor']
data_frame[categorical_columns] = data_frame[categorical_columns].astype('category')
```

```python
float_columns = list(data_frame.iloc[:, 2:-1].columns)
data_frame[float_columns] = data_frame[float_columns].fillna(value=0)
data_frame[float_columns] = data_frame[float_columns].applymap(lambda x: str(x).replace('R$','').replace('.','').replace('%','').replace(',', '.'))
data_frame[float_columns] = data_frame[float_columns].astype('float')
```

```python
indicadores = [
    'Setor',
    'DY (12M)Acumulado',
    'VacânciaFísica',
    'P/VPA',
    'QuantidadeAtivos',
    'Liquidez Diária',
    'PatrimônioLíq.',
]
```

```python
setores = [
    "Hospital",
    "Hotel",
    "Híbrido",
    "Lajes Corporativas",
    "Logística",
    "Outros",
    "Residencial",
    "Shoppings",
    "Títulos e Val. Mob.",
]
```

# Mostrando Oportunidades

```python
def oportunidade_media_setor(dataframe, label_setor='Setor', setor=None):
    media_setor = df_aux.groupby('Setor').agg(['mean', 'std'])
    df_setor = dataframe[dataframe[label_setor].isin([setor])]
    filter = None
    
    if setor == "Hospital":
        filter = \
            (df_setor['Liquidez Diária'] > 1)
    elif setor == "Hotel":
        filter = \
            (df_setor['QuantidadeAtivos'] > 3) & \
            (df_setor['Liquidez Diária'] > 1000) & \
            (df_setor['P/VPA'] < 1.0) & \
            (df_setor['DY (12M)Acumulado'] > media_setor.loc[setor, ('DY (12M)Acumulado', 'mean')])
    elif setor == "Híbrido":
        filter = \
            (df_setor['QuantidadeAtivos'] > 0) & \
            (df_setor['Liquidez Diária'] > 5000) & \
            (df_setor['P/VPA'] < 1.0) & \
            (df_setor['DY (12M)Acumulado'] > media_setor.loc[setor, ('DY (12M)Acumulado', 'mean')])
    elif setor == "Lajes Corporativas":
        filter = \
            (df_setor['QuantidadeAtivos'] > 0) & \
            (df_setor['Liquidez Diária'] > 5000) & \
            (df_setor['P/VPA'] < 1.05) & \
            (df_setor['DY (12M)Acumulado'] > media_setor.loc[setor, ('DY (12M)Acumulado', 'mean')])
    elif setor == "Logística":
        filter = \
            (df_setor['QuantidadeAtivos'] > 0) & \
            (df_setor['Liquidez Diária'] > 4000) & \
            (df_setor['P/VPA'] < 1.5) & \
            (df_setor['DY (12M)Acumulado'] > media_setor.loc[setor, ('DY (12M)Acumulado', 'mean')])
    elif setor == "Outros":
        filter = \
            (df_setor['Liquidez Diária'] > 5000) & \
            (df_setor['P/VPA'] < 1.0) & \
            (df_setor['DY (12M)Acumulado'] > media_setor.loc[setor, ('DY (12M)Acumulado', 'mean')])
    elif setor == "Residencial":
        filter = \
            (df_setor['QuantidadeAtivos'] > 0) & \
            (df_setor['Liquidez Diária'] > 5000) & \
            (df_setor['P/VPA'] < 1.0) & \
            (df_setor['DY (12M)Acumulado'] > media_setor.loc[setor, ('DY (12M)Acumulado', 'mean')])
    elif setor == "Shoppings":
        filter = \
            (df_setor['QuantidadeAtivos'] > 0) & \
            (df_setor['Liquidez Diária'] > 1000) & \
            (df_setor['P/VPA'] < 1.1) & \
            (df_setor['DY (12M)Acumulado'] > media_setor.loc[setor, ('DY (12M)Acumulado', 'mean')])
    elif setor == "Títulos e Val. Mob.":
        filter = \
            (df_setor['Liquidez Diária'] > 5000) & \
            (df_setor['P/VPA'] < 1.05) & \
            (df_setor['DY (12M)Acumulado'] > media_setor.loc[setor, ('DY (12M)Acumulado', 'mean')])
    
    return df_setor[filter]
```

```python
for setor in setores:
    oportunidade = oportunidade_media_setor(data_frame, setor=setor)

    if not oportunidade.empty:
        print(setor)
        oportunidade.sort_values('DY (12M)Acumulado', ascending=False, inplace=True)
        display(oportunidade)
        print('\n')
```
